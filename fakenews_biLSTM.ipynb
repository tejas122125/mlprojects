```python
y_train = df["label"]
y_train.value_counts()
```

```python
messages=x_train.copy()
```

```python
import nltk
import re
from nltk.corpus import stopwords
```

```python
nltk.download('stopwords')
```

```python
messages.reset_index(inplace=True)
```

```python
### Dataset Preprocessing
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
corpus = []
for i in range(0, len(messages)):
    review = re.sub('[^a-zA-Z]', ' ', messages['title'][i])
    review = review.lower()
    review = review.split()

    review = [ps.stem(word) for word in review if not word in stopwords.words('english')]
    review = ' '.join(review)
    corpus.append(review)
```

```python
corpus
```

```python
type(corpus)
```

```python
onehot_repr=[one_hot(words,5000) for words in corpus]
onehot_repr
```

```python
sent_length=20
embedded_docs=pad_sequences(onehot_repr,padding='pre',maxlen=sent_length)

embedded_docs
```

```python
embedding_vector_features=40
model=Sequential()
model.add(Embedding(5000,embedding_vector_features,input_length=sent_length))
model.add(Bidirectional(LSTM(100)))
model.add(Dense(32,activation = 'relu'))
model.add(Dense(1,activation='sigmoid'))
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
print(model.summary())
```

```python

X_final=np.array(embedded_docs)
y_final=np.array(y_train)
```

```python
from tensorflow.keras.callbacks import EarlyStopping

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_final, y_final, test_size=0.33, random_state=42)

early_stopping_callback = EarlyStopping(monitor='val_loss', patience=3, restore_best_weights=True)
```

```python
model.fit(X_train,y_train,validation_data=(X_test,y_test),epochs=10,batch_size=32,callbacks=[early_stopping_callback])
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
```

```python
y_pred=model.predict(X_test)
# print(accuracy_score(y_test,y_pred))
y_pred=np.where(y_pred>=0.5,1,0)
print(accuracy_score(y_test,y_pred))
```
